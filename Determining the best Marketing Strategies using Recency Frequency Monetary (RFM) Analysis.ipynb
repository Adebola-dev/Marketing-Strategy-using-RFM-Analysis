Interpretation of Durbin Watson Test Statistics - Use to test for autocorrelation between the Income and the X variables(type of products)

**Hypothesis Statement***
Ho: There is no correlation among X-Products type residuals
Ha: There is correlation among X-Products type residuals

**Basic for Durbin Watson Test Statistic***
A test statistic of 2 indicates no serial correlation.
The closer the test statistics is to 0, the more evidence of positive serial correlation.
The closer the test statistics is to 4, the more evidence of negative serial correlation.

***Conclusion: Since the Durbin Watson Test statistic is 1.860 closer to 2, then we accept Ho and reject Ha concluding that there is positive serial correlation among the X-Products type residuals.

***RFM Analysis (Recency, Frequency, and Monetary value) for Customer Segmentation- RFM metrics are important indicators of a customer’s behavior because frequency and monetary value affects a customer’s lifetime value, and recency affects retention, a measure of engagement.

***To derive the RFM analysis

There is need to group (NumDealsPurchases, NumWebPurchases, NumCatalogPurchases, NumStorePurchases) together as purchases_col.
There is need to group (MntWines, MntFruits, MntMeatProducts, MntFishProducts, MntSweetProducts, MntGoldProds) together as amount_col

Therefore, the following columns will be called ID, Recency, total purchases and total amount.

Please note***
NumWebVisitMonth was excluded because the customer visited the company's website but no purchases were made.

```python
purchases_cols=['NumDealsPurchases','NumWebPurchases',
                'NumCatalogPurchases','NumStorePurchases']
amount_cols=['MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds']

df['total_purch']=df[purchases_cols].sum(axis=1)
df['total_amt']=df[amount_cols].sum(axis=1)

rfmTable = (df.groupby('ID').agg({'Recency': lambda x: x.sum(),
                                     'total_purch': lambda x: x.sum(),     
                                        'total_amt': lambda x: x.sum()}))

rfmTable.rename(columns={'Recency': 'recency', 
                         'total_purch': 'frequency', 
                         'total_amt': 'monetary_value'}, inplace=True)

rfmTable.head()
```

```python
quantiles = rfmTable.quantile(q=[0.25,0.5,0.75])
quantiles
```

```python
#Send quantiles to a dictionary, easier to use.
quantiles = quantiles.to_dict()

#Column assign to recency, frequency and monetary_value
quantiles
```

```python
#creating the RFM segmentation table
rfmSegmentation = rfmTable
```

```python
#Arguments (x = value, p = recency, monetary_value, frequency, k = quartiles dict) where RClass stands for Recency Class, FMClass stands for Freqency Monetary value class.

def RClass(x,p,d):     
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]: 
        return 3
    else:
        return 4
    
# Arguments (x = value, p = recency, monetary_value, frequency, k = quartiles dict)
def FMClass(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]: 
        return 2
    else:
        return 1
```

```python
#Using the two segementation process: "R" and "FM" Classes
rfmSegmentation['R_Quartile'] = rfmSegmentation['recency'].apply(RClass, args=('recency',quantiles,))
rfmSegmentation['F_Quartile'] = rfmSegmentation['frequency'].apply(FMClass, args=('frequency',quantiles,))
rfmSegmentation['M_Quartile'] = rfmSegmentation['monetary_value'].apply(FMClass, args=('monetary_value',quantiles,))
```

```python
#Table using the Recency Freqency Monetary value with R_Quartile; F_Quartile; M_Quartile
rfmSegmentation['RFMClass'] = rfmSegmentation.R_Quartile.map(str)\
                            + rfmSegmentation.F_Quartile.map(str) \
                            +  rfmSegmentation.M_Quartile.map(str)
rfmSegmentation.head()
```

```python
#Using the two segementation process: "R" and "FM" Classes
rfmSegmentation['R'] = rfmSegmentation['recency'].apply(RClass, args=('recency',quantiles,))
rfmSegmentation['F'] = rfmSegmentation['frequency'].apply(FMClass, args=('frequency',quantiles,))
rfmSegmentation['M'] = rfmSegmentation['monetary_value'].apply(FMClass, args=('monetary_value',quantiles,))
```

```python
#Table using the Recency Freqency Monetary value
rfmSegmentation['RFMClass'] = rfmSegmentation.R.map(str) + rfmSegmentation.F.map(str) + rfmSegmentation.M.map(str)
rfmSegmentation['RFMScores'] = rfmSegmentation[['R','F','M']].sum(axis=1)
rfmSegmentation.head(10)
```

**RFMScores gives information about the customers loyality and engagement. Hence, RFMScores is the sum of RFM. And the lower the RFMScores the more engage or loyal the customer.

**INTERPRETATION:Customer with the 'ID25'{4} is more loyal than customer with 'ID20 an ID13'{11}.

```python
#Assign loyalty level to each customers [Customers Segmentation]
Loyalty_level = ['Platinum','Gold','Silver','Bronze']
Score_cuts = pd.qcut(rfmSegmentation.RFMScores, q = 4, labels = Loyalty_level)
rfmSegmentation['Cust_Segm_Loyalty_Level'] = Score_cuts.values
rfmSegmentation.reset_index().head(10)
```
